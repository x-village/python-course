# 欲求 1~1000 中可以整除11的最大數

```python
n = 0
ans = 0
while n <= 1000:
    if (n % 11 == 0):
        ans = n
    n += 1
print(ans)
```

## 思考一下
### 找最大的數，從1000往1找比較快找到，還是我從1開始往1000找?

```python
n = 1000
flag = True
while n > 0:
    if (n % 11 == 0) and flag:
        print(n)
        flag = False
    n -= 1
```

![](https://i.imgur.com/Tpn39TP.png)

## 明明找10次最大值 990 就找到了，可是程式還是找了1000次！

## break ：中斷其所在的最接近迴圈

![](https://i.imgur.com/Go8XID4.png)


```python
n = 1000
while n > 0:
    if (n % 11 == 0):
        print(n)
        break
    n -= 1
```

# 巢狀迴圈(nested loop) - while


```python
i = 1   # 被乘數
j = 1   # 乘數
while i <= 9:
    while j <= 9:
        print(i, '*', j, '=', i* j, sep='', end='\t')
        j += 1
    print("")
    j = 1
    i += 1
```

## 從1加到100

## for

```python
total = 0
for i in range(101):
    total += i
print(total)
```

## while 

```python
n = 1
total = 0
while n <= 100:
    total += n
    n += 1
print(total)
```

## for with list

## for with string

## break

```python
for i in range(1000,0,-1):
    if (i % 11 == 0):
        print(i)
        break
```

## continue

```python
for i in range(0,11):
    if(i % 2 == 0):
        print(i, "is even.")
        continue
    print(i, "is odd.")
```

## 找出list中的最大值

```python
a = [10, 2, 8, 15, 21, 1]
max = 0
for i in a:
    if max < i:
        max = i
print(max)
```

## 印出99乘法表

```python
for i in range(1, 10):
    for j in range(1, 10):
        print(i, '*', j, '=', i* j, sep = '', end = '\t')
    print("")
```
