# 課堂練習ex2 - 找出空氣最好的地方
## 列入計分
1. 請去資料來源下載 aqi 的 csv 格式
2. aqi.csv 是 7/5 全台灣空氣指數的資料，讀取這個檔案
3. 找出 aqi 最低的數值出現的地區，並且數值是多少
4. 印出答案： XX 空氣最好，AQI 是 XX

[資料來源](https://data.gov.tw/dataset/40448)

# 參考解答（資料會改變，結果有可能不同）

````python
# 解1: 假設最小的 aqi 只有一個地方
import csv


AQI_INDEX = 2
SITE_INDEX = 0


with open('aqi.csv', encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    sites = []
    aqis = []
    is_first_line = True
    
    for row in reader:
        # 考慮有些地區資料缺漏
        if is_first_line or not row[AQI_INDEX]:
            is_first_line = False
            continue
        sites.append(row[SITE_INDEX])
        aqis.append(int(row[AQI_INDEX]))
    
    min_aqi = min(aqis)
    min_aqi_site = sites[aqis.index(min_aqi)]

print('{}空氣最好， AQI是{}'.format(min_aqi_site, min_aqi))
````

````python
# 解2: 考慮到最小的 aqi 值有可能出現在不止一個地方
import csv


AQI_INDEX = 2
SITE_INDEX = 0


with open('aqi.csv', encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    sites = []
    aqis = []
    is_first_line = True
    
    for row in reader:
        # 考慮有些地區資料缺漏
        if is_first_line or not row[AQI_INDEX]:
            is_first_line = False
            continue
        sites.append(row[SITE_INDEX])
        aqis.append(int(row[AQI_INDEX]))
    
    min_aqi = min(aqis)
    # 主要差異在這行, 找出所有出現 min_aqi 的地區
    # reference: https://docs.python.org/3/tutorial/datastructures.html#list-comprehensions
    min_aqi_sites = [x for i, x in enumerate(sites) if aqis[i] == min_aqi]

print('{}空氣最好， AQI是{}'.format(','.join(min_aqi_sites), min_aqi))
````

# 課程練習ex3 - Pretty print Open data: 獨立音樂展演
## 列入計分

1. 下載獨立音樂展演公開資料的 JSON 格式
